```python
data_bitcoin = df_bitcoin['data']['quotes']

# Open the file in write mode
file = open('data_bitcoin.csv', 'w')

for  val in data_bitcoin:
    temp = val['quote']
    open_p = temp['open']
    close_p = temp['close']
    high_p = temp['high']
    low_p =  temp['low']
    volume = temp['volume']
    market_cap = temp['marketCap']
    timestamp = temp['timestamp']
    file.write(str(open_p) + ',' + str(close_p) + ',' + str(high_p) + ',' + str(low_p) + ',' + str(volume) + ',' + str(market_cap) + ',' + str(timestamp) + "\n")

# Close the file
file.close()
```
